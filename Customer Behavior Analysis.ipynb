```python
df['review_rating'] = df.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
data = []

for index, row in df.iterrows():
    data.append((
        int(row["customer_id"]),
        int(row["age"]),
        row["gender"],
        row["item_purchased"],
        row["category"],
        float(row["purchase_amount"]),
        row["location"],
        row["size"],
        row["color"],
        row["season"],
        float(row["review_rating"]),
        row["subscription_status"],
        row["shipping_type"],
        row["discount_applied"],
        int(row["previous_purchases"]),
        row["payment_method"],
        row["frequency_of_purchases"]
    ))
sql = """
INSERT INTO customers
(customer_id, age, gender, item_purchased, category,
 purchase_amount, location, size, color, season,
 review_rating, subscription_status, shipping_type,
 discount_applied, previous_purchases, payment_method,
 frequency_of_purchases)
VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
"""

cursor.executemany(sql, data)
conn.commit()

print("Data inserted successfully")
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("customer_shopping_behavior.csv")

fig = plt.figure(figsize=(18, 12))

# KPIs
fig.text(0.1, 0.93, f"Total Customers: {df.customer_id.nunique()}", fontsize=12)
fig.text(0.35, 0.93, f"Total Revenue: {df.purchase_amount.sum()}", fontsize=12)
fig.text(0.6, 0.93, f"Avg Purchase: {round(df.purchase_amount.mean(),2)}", fontsize=12)
fig.text(0.8, 0.93, f"Top Category: {df.category.mode()[0]}", fontsize=12)

# Charts
plt.subplot(2,3,1)
df.groupby("category")["purchase_amount"].sum().plot(kind="bar")
plt.title("Purchase by Category")

plt.subplot(2,3,2)
df["gender"].value_counts().plot(kind="bar")
plt.title("Purchases by Gender")

plt.subplot(2,3,3)
df["payment_method"].value_counts().plot(kind="bar")
plt.title("Payment Methods")

plt.subplot(2,3,4)
df.groupby("season")["purchase_amount"].sum().plot(kind="bar")
plt.title("Season-wise Sales")

plt.subplot(2,3,5)
df["discount_applied"].value_counts().plot(kind="bar")
plt.title("Discount Impact")

plt.subplot(2,3,6)
plt.hist(df["age"], bins=10)
plt.title("Age Distribution")

plt.tight_layout(rect=[0,0,1,0.9])
```
